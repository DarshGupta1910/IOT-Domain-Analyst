### Correcting the DataSet 

```python
data.isnull().sum()
```

```python
Students = data.drop(['race/ethnicity', 'parental level of education'], axis=1)
```

```python
data['math score']=(data['math score']-(data['math score'].mean()))/data['math score'].std()
```

```python
data['reading score']=(data['reading score']-(data['reading score'].min()))/((data['reading score'].max())-(data['reading score'].min()))
```
